```python
import pandas as pd
df = pd.read_csv('/content/IMDB Dataset.csv', on_bad_lines='skip')  # Skips problematic lines
df.head()
```

#`Text Cleaning`
- Sample 10000 rows
- Remove html tags
- Remove special characters
- Converting every thing to lower case
- Removing Stop words
- Stemming

```python
# Text cleaning
# df['review'].str.replace('<br','').str.replace('/> />','')[0]
```

```python
df['review']=df['review'].str.replace(r'<br\s*/?>', '', regex=True)
```

```python
df['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
df.head()
```

```python
df['review']=df['review'].apply(lambda x:x.lower())
```

```python
def removing_special_words(text):
  x=''
  for i in text:
    if i.isalnum():
      x=x+i
    else:
      x=x+' '
  return  x
```

```python
removing_special_words(' th%e @ classic use of the word.it is called oz as that is the nickname given to the oswald maximum security state penitentary. it focuses mainly on emerald city, an experimental section of the prison where all the cells have glass fronts and face inwards, so privacy is not high on the agenda. em city is home to many..aryans, muslims, gangstas, latinos, christians, italians, irish and more....so scuffles, death stares, dodgy dealings and shady agreements are never far away.i would say the main appeal of the show is due to the fact that it goes where other shows wouldnt dare. forget pretty pictures painted for mainstream audiences, f')
```

```python
df['review']=df['review'].apply(removing_special_words)
```

```python
import nltk

# Download the stopwords dataset
nltk.download('stopwords')

# Verify stopwords are now accessible
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
def removing_stop_words(text):
  x=[]
  for i in text.split():
    if i not in stopwords.words('english'):
      x.append(i)
  y=x[:]
  x.clear()
  return y
```

```python
df['review']=df['review'].apply(removing_stop_words)
```

```python
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
y=[]
def stem_words(text):
  for i in text:
    y.append(ps.stem(i))
  z=y[:]
  y.clear()
  return z
```

```python
df['review']=df['review'].apply(stem_words)
```

```python
df.head()
```

```python
def join(input_list):
  return " ".join(input_list)
```

```python
df['review']=df['review'].apply(join)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
X=cv.fit_transform(df['review']).toarray()
```

```python
X.shape
```

```python
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
clf1=GaussianNB()
clf2=MultinomialNB()
clf3=BernoulliNB()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
clf3.fit(X_train,y_train)
```

```python
y_pred1=clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
y_pred3=clf3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Gaussian",accuracy_score(y_test,y_pred1))
print("Multinomial",accuracy_score(y_test,y_pred2))
print("Bernaulli",accuracy_score(y_test,y_pred3))
```
