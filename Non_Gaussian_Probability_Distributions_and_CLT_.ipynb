#`Test for normal distribution`

```python
df=df.dropna()
```

```python
sns.kdeplot(df['Age'])
```

```python
# Drop missing values in 'Age' column
age_data = df['Age'].dropna()

# Perform Shapiro-Wilk test for normality
shapiro_stat, shapiro_p = stats.shapiro(age_data)

# Perform Anderson-Darling test for normality
anderson_result = stats.anderson(age_data, dist='norm')

# Extract test statistic and critical values
anderson_statistic = anderson_result.statistic
anderson_critical_values = anderson_result.critical_values
anderson_significance_levels = anderson_result.significance_level

# Output results
normality_tests = {
    "Shapiro-Wilk Test": {"Statistic": shapiro_stat, "p-value": shapiro_p},
    "Anderson-Darling Test": {
        "Statistic": anderson_statistic,
        "Critical Values": anderson_critical_values.tolist(),
        "Significance Levels": anderson_significance_levels.tolist(),
    },
}

normality_tests
```

```python
if shapiro_p >=0.05:
  print('Data is  normal dist')
else:
  print('not normally dist.')
```

```python
if anderson_statistic <=0.771:
  print('Data is  normal dist')
else:
  print('not normally dist.')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 2  # Shape parameter (smaller alpha → heavier tail)
scale = 5000  # Minimum city population

# Generate city populations using a Pareto distribution
city_populations = (np.random.pareto(alpha, 1000) + 1) * scale

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(city_populations, bins=50, density=True, alpha=0.6, color='b', label="Histogram")
sns.kdeplot(city_populations, label="kdeplot")
# Calculate theoretical Pareto PDF
x = np.linspace(min(city_populations), max(city_populations), 100)
y = (alpha * scale**alpha) / (x ** (alpha + 1))  # Pareto PDF formula
plt.figure(figsize=(8, 5))
plt.plot(x,y)
# Plot log-log Pareto distribution
plt.figure(figsize=(8, 5))
plt.plot(np.log(x), np.log(y), label="Pareto PDF", color='r')

# Labels
plt.xlabel("log(City Population)")
plt.ylabel("log(Probability Density)")
plt.legend()
plt.show()
```

```python
for col in X_train.columns:
  stats.probplot(X_train[col],dist='norm',plot=plt)
  plt.title(f'columns-{col}')
  plt.show()
```

```python
def transformer(transform):
  num_cols = X_train.columns
  transformer=ColumnTransformer([('power',PowerTransformer(method=transform),X_train.columns)],remainder='passthrough')
  X_train_transformer=transformer.fit_transform(X_train+1e-9)
  X_test_transformer=transformer.transform(X_test+1e-9)
  kf = KFold(n_splits=10, shuffle=True, random_state=42)
  cv_scores_train = cross_val_score(LinearRegression(),X_train, y_train, cv=kf, scoring='r2')
  cv_scores_test = cross_val_score(LinearRegression(),X_test, y_test, cv=kf, scoring='r2')
  print(f"Mean CV Accuracy without trans: {np.mean(cv_scores_train):.4f}")
  print(f"Mean CV Accuracy without trans: {np.mean(cv_scores_test):.4f}")
  cv_scores_train = cross_val_score(LinearRegression(),X_train_transformer, y_train, cv=kf, scoring='r2')
  cv_scores_test = cross_val_score(LinearRegression(),X_test_transformer, y_test, cv=kf, scoring='r2')
  print(f"Mean CV Accuracy: {np.mean(cv_scores_train):.4f}")
  print(f"Mean CV Accuracy: {np.mean(cv_scores_test):.4f}")
  plt.figure(figsize=(10, 5 * len(num_cols)))
  for i, col in enumerate(num_cols):
    plt.subplot(len(num_cols), 2, 2 * i + 1)
    stats.probplot(X_train[col], dist="norm", plot=plt)
    plt.title(f"Before Transformation - {col}")

    plt.subplot(len(num_cols), 2, 2 * i + 2)
    stats.probplot(X_train_transformer[:, i], dist="norm", plot=plt)
    plt.title(f"After Transformation - {col}")

  plt.tight_layout()
  plt.show()
```
