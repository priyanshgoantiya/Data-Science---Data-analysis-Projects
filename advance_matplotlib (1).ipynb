```python
##  subplot with annotation
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

ax[0].scatter(sample_df['avg'], sample_df['strike_rate'], label='Players', marker='o', color='black')
for i in range(sample_df.shape[0]):
  ax[0].annotate(sample_df['batter'].values[i],(sample_df['avg'].values[i],sample_df['strike_rate'].values[i]),textcoords="offset points", xytext=(0,10), ha='center')
ax[0].set_title('Batters Avg vs Strike Rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')
ax[0].legend()

ax[1].scatter(sample_df['runs'], sample_df['strike_rate'], label='Players', marker='+', color='blue')

for i in range(sample_df.shape[0]):
  ax[1].annotate(sample_df['batter'].values[i],(sample_df['runs'].values[i],sample_df['strike_rate'].values[i]),
                             textcoords="offset points", xytext=(0,10), ha='right')
ax[1].set_title('Runs vs Strike Rate')
ax[1].set_xlabel('Runs')
ax[1].set_ylabel('Strike Rate')
ax[1].legend()


ax[2].scatter(sample_df['runs'], sample_df['avg'], label='Players', marker='s', color='green')
for i in range(sample_df.shape[0]):
  ax[2].annotate(sample_df['batter'].values[i],(sample_df['runs'].values[i],sample_df['avg'].values[i]),textcoords="offset points", xytext=(0,10), ha='center')
ax[2].set_title('Runs vs Batters Avg')
ax[2].set_xlabel('Runs')
ax[2].set_ylabel('Average')
ax[2].legend()

plt.tight_layout()
plt.show()
```

## `3D Surface plot`

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
xx.shape
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
T=np.sin(xx)+np.cos(yy)
T.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,T,cmap='viridis')
fig.colorbar(p)
```

```python
t=np.tan(xx)+(1/np.tan(yy))
t.shape
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,t,cmap='viridis')
fig.colorbar(p)
```

```python
df1= pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df1.head()
```

```python
## pie chart
df1[['match1']].plot(kind='pie',labels=df1['batsman'].values,subplots=True,figsize=(10,8),autopct='%0.2f%%')
```

```python
## multiple pie chart
df1[['match1','match2','match3']].plot(kind='pie',labels=df1['batsman'].values,subplots=True,figsize=(20,8),autopct='%0.2f%%',shadow=True,explode=[0,0,0.2,0,0,0])
```

```python
## by using subplots
fig,ax=plt.subplots(1,3,figsize=(20,8))
df1['match1'].plot(kind='pie',labels=df1['batsman'].values,subplots=True,autopct='%0.2f%%',shadow=True,ax=ax[0],explode=[0,0,0.2,0,0,0])
df1['match2'].plot(kind='pie',labels=df1['batsman'].values,subplots=True,autopct='%0.2f%%',shadow=True,ax=ax[1],explode=[0,0,0.2,0,0,0])
df1['match3'].plot(kind='pie',labels=df1['batsman'].values,subplots=True,autopct='%0.2f%%',shadow=True,ax=ax[2],explode=[0,0,0.2,0,0,0])
plt.tight_layout()
plt.show()
```
