```python
## from a pandas dataframe
df=pd.read_csv('/content/sharma-kohli.csv')
df
```

```python
plt.plot(df['index'],df['V Kohli'])
```

```python
## Multiple plotting  and title and  label
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
plt.title('Rohit sharma vs Virat Kohli career comaprison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
## colors(hex) and line(width and style) and marker(size)
plt.plot(df['index'],df['V Kohli'],'D-b',linewidth=2,markersize=4)
plt.plot(df['index'],df['RG Sharma'],'s--g',linewidth=2,markersize=4)
plt.title('Rohit sharma vs Virat Kohli career comaprison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
## add legend -> location and add grid
plt.plot(df['index'],df['V Kohli'],'D-b',linewidth=2,markersize=6,label='Virat Kohli')
plt.plot(df['index'],df['RG Sharma'],'s--g',linewidth=2,markersize=4,label='Rohit Sharma')
plt.title('Rohit sharma vs Virat Kohli career comaprison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
plt.legend()
plt.show()
```

```python
df=pd.read_csv('/content/batsman_season_record.csv')
df.head()
```

```python
## plotting multiple bars ina barplot using  matplotlib
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='black')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='red')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
## stacked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015']+df['2016']),label='2017')
plt.legend()
plt.show()
```
